```python
from ucimlrepo import fetch_ucirepo
import plotly.express as px
import random as rd
from sklearn.model_selection import train_test_split
```

```python
wine_quality = fetch_ucirepo(name='Wine Quality')
wine_data = wine_quality['data']['original']
```

```python
seed = 42 #if it must be 42 :P

wine_te = wine_data.sample(frac=0.8, random_state=42) 
wine_X_te = wine_te.drop(columns = ['quality'])
wine_y_te = wine_te['quality']

wine_X = wine_data.drop(columns = ['quality'])
wine_y = wine_data['quality']

wine_X_tr, wine_X_val, wine_y_tr, wine_y_val = train_test_split(wine_X, wine_y,
                                                                  test_size=0.25, random_state=seed, shuffle=True)
print(wine_X_tr.shape, wine_X_val.shape, wine_y_tr.shape, wine_y_val.shape)
```
