```python
from ucimlrepo import fetch_ucirepo
import plotly.express as px
import plotly.io as pio
```

```python
wine_quality = fetch_ucirepo(name='Wine Quality')
wine_quality_data = wine_quality['data']['original']
wine_quality_data
```

```python
fig = px.histogram(wine_quality_data, x='quality')
fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
pio.write_image(fig, "images/wine_quality.svg")
fig.show()
```

```python
fig = px.box(wine_quality_data, x='quality', y='color')
fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
pio.write_image(fig, "images/wine_color.svg")
fig.show()
```

```python
features = [col for col in wine_quality_data.columns if col not in ['quality', 'color']]

for feature in features:
    fig = px.box(
        wine_quality_data,
        x='quality',
        y=feature,
        color='color',
        color_discrete_map={'red': 'red', 'white': 'gray'}
    )
    fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
    pio.write_image(fig, f"images/wine_{feature}.svg")
    fig.show()
```

```python
fig = px.imshow(
    wine_quality_data.drop("color", axis=1).corr(method='pearson').round(2),
    text_auto=True,
    width=900,
    height=750
)
fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
fig.update_traces(textfont=dict(size=14))
pio.write_image(fig, "images/correlation.svg")
fig.show()
```
