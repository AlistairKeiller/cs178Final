```python
import seaborn as sns
from ucimlrepo import fetch_ucirepo
from datasets import load_dataset
from matplotlib import pyplot as plt
sns.set_style("darkgrid")
```

```python
def plot_data(data):
    for variable in data.columns:
        if data[variable].dtype in ['float64', 'int64']:
            discrete = data[variable].dtype == 'int64' and len(data[variable].unique()) < 20
            sns.histplot(data[variable], kde=not discrete, discrete=discrete)
        else:
            ax = sns.countplot(x=data[variable])
            if len(data[variable].dropna().unique()) > 10:
                ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        plt.tight_layout()
        plt.show()
    sns.heatmap(data.select_dtypes(include=['number']).corr())
    plt.show()
    sns.pairplot(data)
    plt.show()
```

```python
wine_quality = fetch_ucirepo(name='Wine Quality')
wine_quality_data = wine_quality['data']['original']
plot_data(wine_quality_data)
```
